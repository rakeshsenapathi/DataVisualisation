```python
#Importing seaborn, matplotlib for visualisation
import seaborn as sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20, 10)
```

```python
#Read the dataset 
df = pd.read_csv("https://raw.githubusercontent.com/rakeshsenapathi/DataVisualisation/master/data.csv",encoding = "ISO-8859-1")
```

```python
#Check for the datatypes of the features in the dataset
df.dtypes
```

```python
# Data Cleaning
# To see if there are any NaN values in the data set
df.isnull().values.any()
```

```python
df.stn_code.head().T
```

```python
#Starting with station_code which is of object data type initially
#Replacing null values with 0
df.stn_code = df.stn_code.fillna(0)
```

```python
#There are also values with string 'SAMP' as the column value
df.stn_code = df.stn_code.replace('SAMP',0)
df.dtypes
```

```python
#Converting the object data type of station code to integer
df.stn_code = df.stn_code.astype('int64',errors = "raise")
#Converted the data type to int64
```

```python
#Parsing the date object
df.date = df.date.fillna(0)
df['date'] = pd.to_datetime(df['date'])
```
